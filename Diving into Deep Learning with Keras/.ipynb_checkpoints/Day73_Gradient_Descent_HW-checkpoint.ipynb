使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
precision = 0.000001
max_iters = 10000
df = lambda x: 2 * (x+5)
```

```python
'''
需要執行初始值設定, 下列三個
# Learning rate
#iteration counter
#Step size  
'''
'''
while previous_step_size > precision and iters < max_iters:

  算法迭代更新

print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
'''

lr_list = [0.1, 0.0001]

iters_history = [[],[]]
x_history = [[],[]]

for idx in range(len(lr_list)):
    previous_step_size = 1
    iters = 1
    cur_x = 3
    
    iters_history[idx].append(previous_step_size)
    x_history[idx].append(cur_x)
    
    while (previous_step_size > precision ) and (iters < max_iters):
        prev_x = cur_x
        cur_x -= df(cur_x) * lr_list[idx]
        previous_step_size = abs(cur_x - prev_x)
        iters += 1
        iters_history[idx].append(iters)
        x_history[idx].append(cur_x)
        print(f'prev_x: {prev_x}, cur_x: {cur_x}')
        print(f'iter: {iters},  previous_step_size: {previous_step_size}')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#適用於 Jupyter Notebook, 宣告直接在cell 內印出執行結果

for idx in range(len(lr_list)):
    print(f'Learning rate: {lr_list[idx]}')
    plt.plot(iters_history[idx], x_history[idx], 'o-', ms=3, lw=1.5, color='black')
    plt.xlabel(r'$iters$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.show()
```

## Ans
在此次遞迴中，使用0.1的learning rate在大約30次後即收斂，而0.0001的learning rate直到一萬場也尚未收斂。
