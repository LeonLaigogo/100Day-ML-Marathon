```python
import os
import keras

# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
def build_mlp(input_shape, output_units=10, num_neurons=[512, 256, 128]):
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_units in enumerate(num_neurons):
        if i == 0:
            x = keras.layers.Dense(units=n_units, activation="relu", name="hidden_layer"+str(i+1))(input_layer)
        else:
            x = keras.layers.Dense(units=n_units, activation="relu", name="hidden_layer"+str(i+1))(x)
    
    out = keras.layers.Dense(units=output_units, activation="softmax", name="output")(x)
    
    model = keras.models.Model(inputs=[input_layer], outputs=[out])
    return model
```

```python
## 超參數設定
LEARNING_RATE = [0.01, 0.005, 0.001] 
EPOCHS = 20
BATCH_SIZE = 256
MOMENTUM = 0.95 
# optimizer = ['SGD', 'RMSprop', 'AdaGrad', 'Adam']
```

```python
# Using Adam
results = {}
keras.backend.clear_session() 
model = build_mlp(input_shape=x_train.shape[1:])
model.summary()
optimizer = keras.optimizers.Adam()
model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer=optimizer)

model.fit(x_train, y_train, 
          epochs=EPOCHS, 
          batch_size=BATCH_SIZE, 
          validation_data=(x_test, y_test), 
          shuffle=True)

# Collect results
train_loss = model.history.history["loss"]
valid_loss = model.history.history["val_loss"]
train_acc = model.history.history["acc"]
valid_acc = model.history.history["val_acc"]

exp_name_tag = f"Adam result"
results[exp_name_tag] = {'train-loss': train_loss,
                         'valid-loss': valid_loss,
                         'train-acc': train_acc,
                         'valid-acc': valid_acc}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,6))
for i, cond in enumerate(results.keys()):
#     plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond, color=color_bar[i])
#     plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond, color=color_bar[i])
    plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond)
    plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond)
plt.title("Loss")
plt.legend()
plt.show()
plt.figure(figsize=(8,6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train-acc'])),results[cond]['train-acc'], '-', label=cond)
    plt.plot(range(len(results[cond]['valid-acc'])),results[cond]['valid-acc'], '--', label=cond)
plt.title("Accuracy")
plt.legend(loc='best')
plt.show()
```

```python
#Using SGD to choose hyper parameters
'''
 SGD(隨機梯度下降) - Arguments
lr: float >= 0. Learning rate.
momentum: float >= 0. Parameter that accelerates SGD in the relevant direction and dampens oscillations.
decay: float >= 0. Learning rate decay over each update.
nesterov: boolean. Whether to apply Nesterov momentum.
'''
LEARNING_RATE = [0.01, 0.001] 
momentum = [0.95, 0.75]
decay = [0.1, 0.3]
nesterov = [True, False]

for lr in LEARNING_RATE:
    for mom in momentum:
        for de in decay:
            for nes in nesterov:
                print("Experiment with Learning Rate = %.6f" % (lr))
                print("Experiment with MOMENTUM = %.6f" % (mom))
                print("Experiment with decay = %.6f" % (de))
                print(f"Using Nesterov: {nes}")
                
                keras.backend.clear_session() # 把舊的 Graph 清掉

                model = build_mlp(input_shape=x_train.shape[1:])
                model.summary()
                optimizer = keras.optimizers.SGD(lr=lr, momentum=mom, decay=de, nesterov=nes)
                model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer=optimizer)
                
                model.fit(x_train, y_train, 
                          epochs=EPOCHS, 
                          batch_size=BATCH_SIZE, 
                          validation_data=(x_test, y_test), 
                          shuffle=True)

                # Collect results
                train_loss = model.history.history["loss"]
                valid_loss = model.history.history["val_loss"]
                train_acc = model.history.history["acc"]
                valid_acc = model.history.history["val_acc"]

                exp_name_tag = f"MOMENTUM-{mom}-Nes-{nes}"
                results[exp_name_tag] = {'train-loss': train_loss,
                                         'valid-loss': valid_loss,
                                         'train-acc': train_acc,
                                         'valid-acc': valid_acc}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,16))
for i, cond in enumerate(results.keys()):
#     plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond, color=color_bar[i])
#     plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond, color=color_bar[i])
    plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond)
    plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond)
plt.title("Loss")
plt.legend()
plt.show()
```
