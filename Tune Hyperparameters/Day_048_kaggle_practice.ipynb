```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
from IPython.display import display
train_data = pd.read_csv('train.csv', header=None)
test_data = pd.read_csv('test.csv', header=None)
trainLabels = pd.read_csv('trainLabels.csv', header=None)
display(train_data[0:10])
display(test_data[0:10])
display(trainLabels[0:10])
```

```python
X = train_data
y = trainLabels
```

```python
test_size = 0.25
random_state = 1234
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
clf = GradientBoostingClassifier(random_state=random_state)

clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

print(f'bench line score: {metrics.accuracy_score(y_test, y_pred)}')
```

```python
n_estimators = [100, 200, 300]
max_depth = [1, 3, 5]
param_grid = dict(n_estimators = n_estimators, max_depth = max_depth)

grid_search = GridSearchCV(clf, param_grid, scoring='accuracy', n_jobs=-1)

grid_result = grid_search.fit(x_train, y_train)

print('Best score %f using %s' % (grid_result.best_score_, grid_result.best_params_))
```

```python
grid_result.best_params_
```

```python
clf_best_params = GradientBoostingClassifier(n_estimators = grid_result.best_params_['n_estimators'],
                                             max_depth = grid_result.best_params_['max_depth'])

clf_best_params.fit(x_train, y_train)

y_pred = clf_best_params.predict(x_test)

print(f'clf_best_params socre: {metrics.accuracy_score(y_test, y_pred)}')
```

```python
mms = MinMaxScaler()
nor = Normalizer()
X_mms = mms.fit_transform(X)
X_norm = nor.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_mms, y, test_size=test_size, random_state=random_state)
clf_best_params = GradientBoostingClassifier(n_estimators = grid_result.best_params_['n_estimators'],
                                             max_depth = grid_result.best_params_['max_depth'])

clf_best_params.fit(x_train, y_train)

y_pred = clf_best_params.predict(x_test)

print(f'clf_best_params with MinMaxScaler socre: {metrics.accuracy_score(y_test, y_pred)}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_norm, y, test_size=test_size, random_state=random_state)
clf_best_params = GradientBoostingClassifier(n_estimators = grid_result.best_params_['n_estimators'],
                                             max_depth = grid_result.best_params_['max_depth'])

clf_best_params.fit(x_train, y_train)

y_pred = clf_best_params.predict(x_test)

print(f'clf_best_params with MinMaxScaler socre: {metrics.accuracy_score(y_test, y_pred)}')
```

```python
final_result = pd.DataFrame(clf_best_params.predict(nor.fit_transform(test_data)))

final_result.index += 1

final_result.columns = ['Solution']
final_result['Id'] = np.arange(1, final_result.shape[0]+1)
final_result = final_result[['Id', 'Solution']]
print(final_result)
```
